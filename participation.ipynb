```python
import math
class Point: # class definition, starts with keyword "class"
    '''Creates a point on a coordinate plane with values x and y.'''

    def __init__(self, x, y): # method called when an object is created
        '''Defines x and y variables'''
        self.x = x
        self.y = y

    def __str__(self): # method called when an object is printed
        return "Point(%s,%s)"%(self.x, self.y)
    import math
    def manhattanDistance(self, other):
      dx = abs(self.x - other.x)
      dy = abs(self.y - other.y)
      return dx + dy
    def distance(self, other):
      dx = abs(self.x - other.x)
      dy = abs(self.y - other.y)
      return math.sqrt(dx**2 + dy**2)
    def manhattanDistance(self, other):
      return abs(self.x - other.x) + abs(self.y - other.y)


p=Point(0,0)
q=Point(3,4)
print(p.x)
print(p.y)
print(p)
print(q)
print(p.manhattanDistance(q))
```

```python
class Circle:
  def __init__(self, pt, r):
    self.center = pt
    self.radius = r
    self.area = math.pi * (r ** 2)
    self.perimeter = math.pi * 2 * r

  def intersect(self, other):
    return self.center.distance(other.center) <= (self.radius + other.radius)

```

```python
class clock_iterator:
  def __init__(self, start):
    self.cur = start
  def __iter__(self):
    return self
  def __next__(self):
    ret = self.cur
    if self.cur == 23:
      self.cur = 0
    else:
      self.cur += 1
    return ret
```

```python
def fib():
  x = 0
  y = 1
  while True:
    yield x
    x, y = y, x + y

i = 0
for i in fib():
  print(i)
  if i > 5:
    break
  i += 1
```

```python
def stopwatch():
  hr, min, sec = 0, 0, 0
  while True:
    yield hr, min, sec
    sec += 1
    if sec == 60:
      sec = 0
      min += 1
    if min == 60:
      min = 0
      hr += 1
    if hr == 24:
      hr = 0

sw = stopwatch()
for i in range(1000000):
  hours, mins, secs = next(sw)
print(hours,mins,secs)
```
