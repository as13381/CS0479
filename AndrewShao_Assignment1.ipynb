Question 1

```python
# helper function to determine if integer n is prime
def isPrime(n):
  for x in range(2, int(n ** 0.5) + 1):
    if n % x == 0:
      return False
  return n != 1

def prime_factors(n):
  i = 2
  pfs = []
  while i < n:
    # if n is divisible by i, i must be a prime factor
    if n % i == 0:
      n = int(n / i)
      pfs.append(i)
      # if n divided by i is prime, quotient is last prime factor
      if isPrime(n):
        pfs.append(n)
        return pfs
      i = 2
    else:
      i += 1
  return([n])

# prime_factors(100)
```

Question 2

```python
def prime_factors_r(n):
  if isPrime(n):
    return [n]
  # find two integers which product is n
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      # the prime factors of n are the prime factors of both numbers whose product is n
      return prime_factors_r(i) + prime_factors_r(int(n / i))

# prime_factors_r(100)
```

```python
def Venn(m, n):
  prime_m = set(prime_factors(m))
  prime_n = set(prime_factors(n))
  return [prime_m.difference(prime_n), prime_n.difference(prime_m), prime_m.intersection(prime_n)]

# print(prime_factors(60))
# print(prime_factors(12))
# print(Venn(23, 46))
```

```python
def same_primefs(m, n):
  # if two numbers have the same set of prime factors, the number of elements in set differences should be 0 both ways
  if len(Venn(m, n)[0]) +  len(Venn(m, n)[1]):
    return(False)
  return(True)

# same_primefs(40, 100)
```

```python
def encrypt(str1, str2):
  retstr = ""
  for i in range(len(str1)):
    # rotate corresponding character in str1 by appropriate str2 number
    # second argument of rotate is str2 nums list multiplied to appropriate size greater than or equal to length of str1
    retstr += rotate(str1[i], (letter_nums(str2) * (len(str1) // len(str2) + 1))[i])
  return retstr

# encrypt("zebra", "ab")
```

```python
def trapError(func):
  def wrapper(n):
    # if argument is a string, return None
    if type(n) == str:
      return(None)
    return func(n)
  return wrapper

@trapError
def square(n):
  return n ** 2

# print(square("8"))
# print(square(8.0))
```
