1.

```python
import pandas as pd
import numpy as np

titles = pd.read_csv('data/netflix_titles.csv')

titles.head()
```

2.

```python
cast_work = titles.filter(['show_id', 'cast'])
cast_work['cast'] = cast_work['cast'].str.split(',')

cast_work.head()
```

3.

```python
c2 = cast_work.explode('cast')
c2.columns = ['show_id', 'Actor']

c2.head()
```

```python
actors = pd.DataFrame({'Actor': np.sort(c2['Actor'].dropna().unique())})
actors['Actor_ID'] = np.arange(1, actors.shape[0] + 1)

actors.head()
```

4.

```python
show_actor = c2.merge(actors, on='Actor').drop('Actor', axis=1)

show_actor.head()
```

5.

```python
movie_runtimes = titles[titles['type'] == 'Movie'].filter(['show_id', 'duration'])
movie_runtimes['duration'] = pd.to_numeric(movie_runtimes['duration'].str.split(' ').str.get(0))
movie_runtimes.columns = ['show_id', 'runtime']

movie_runtimes.head()
```

6.

```python
tv_n_seasons = titles[titles['type'] == 'TV Show'].filter(['show_id', 'duration'])
tv_n_seasons['duration'] = pd.to_numeric(tv_n_seasons['duration'].str.split(' ').str.get(0))
tv_n_seasons.columns = ['show_id', 'n_seasons']

tv_n_seasons.head()
```

7.

```python
genres_work = titles.filter(['show_id', 'listed_in'])
genres_work['listed_in'] = genres_work['listed_in'].str.split(',')

genres_work.head()
```

8.

```python
g2 = genres_work.explode('listed_in')
g2.columns = ['show_id', 'genre']

g2.head()
```

9.


```python
genres = pd.DataFrame({'genre': np.sort(g2['genre'].dropna().unique())})
genres['genre_id'] = np.arange(1, genres.shape[0] + 1)

genres.head()
```

10.

```python
show_genre = g2.merge(genres, on='genre').drop('genre', axis=1)

show_genre.head()
```

11.

```python
titles.drop(['duration', 'cast', 'listed_in'], axis=1, inplace=True)

titles.head()
```
