Question 2

```python
dim = 9
arr0 = np.zeros((dim, dim))
for i in range(dim // 4 + 1):
  for j in range(0 + 2 * i, dim - 2 * i):
    arr0[0 + 2 * i][j] = 1
    arr0[dim - 1 - 2 * i][j] = 1
    arr0[j][0 + 2 * i] = 1
    arr0[j][dim - 1 - 2 * i] = 1

print(arr0)

arr1 = np.zeros((3, 3))
arr1[1][1] = 1
arr1 = np.pad(arr1, 1, mode='constant', constant_values=1)
arr1 = np.pad(arr1, 1, mode='constant', constant_values=0)
arr1 = np.pad(arr1, 1, mode='constant', constant_values=1)

print(arr1)
```

```python
arr = np.arange(1, 101)
arr[0] = 0
for i in [2, 3, 5, 7]:
  arr[2 * i - 1::i] = 0
arr = arr[arr > 0]

print(arr)
```
